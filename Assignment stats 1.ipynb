## Problem Statement 3:

The number of times I go to the gym in weekdays, are given below along with its associated probability:

x = 0, 1, 2, 3, 4, 5

f(x) = 0.09, 0.15, 0.40, 0.25, 0.10, 0.01

Calculate the mean no. of workouts in a week. Also evaluate the variance involved in  it.

```python
x = [0, 1, 2, 3, 4, 5]
fx = [0.09, 0.15, 0.40, 0.25, 0.10, 0.01]
mean_x = np.dot(x, fx)
x_2 = list(map(lambda x:x**2, x))
var_x = np.dot(x_2, fx) - mean_x**2
print('Mean no. of workouts in a week:', mean_x)
print('Variance:', var_x)
```

## Problem Statement 5:

A company manufactures LED bulbs with a faulty rate of 30%. If I randomly select 6 chosen LEDs, what is the probability of having 2 faulty LEDs in my sample? Calculate the average value of this process. Also evaluate the standard deviation associated with it.

```python
from scipy.stats import binom

# number of trials is n and p is probability of success for each trial
n = 6
p = 0.3
# number of successful trials is x
x = 2

probab = binom.pmf(x, n, p)
print("Probability of having 2 faulty LEDs := ", round(probab, 3))
mean, var = binom.stats(n, p)
print('Average value of the process := ', mean)
print('Standard deviation := ', np.sqrt(var))
```

## Problem Statement 7:

Customers arrive at a rate of 72 per hour to my shop. What is the probability of ùëò customers arriving in 4 minutes? a) 5 customers, b) not more than 3 customers, c) more than 3 customers. Give a pictorial representation of the same to validate your
answer.

```python
from scipy.stats import poisson
# Rate of customers arriving per hour is 72.
# Rate of customers arriving in 4 minutes is mu
mu = 72*4/60
# a) 5 customers. P(x=5)
x = 5
prob1 = poisson.pmf(x, mu)
print('Probability of 5 customers arriving in 4 mins:', round(prob1, 4))
# b) Not more than 3 customers. P(x<=3)
x = 3
prob2 = poisson.cdf(x, mu)
print('Probability of not more than 3 customers arriving in 4 mins:', round(prob2, 4))
# c) More than 3 customers. P(x>3) = 1 - P(x<=3)
prob3 = 1 - prob2
print('Probability of more than 3 customers arriving in 4 mins:', round(prob3, 4))

x_values = list(range(12))
prob_dist = [poisson.pmf(x_value, mu) for x_value in x_values]
cum_dist = [poisson.cdf(x_value, mu) for x_value in x_values]

fig, axes = plt.subplots(1, 2, figsize=(16, 4))
axes[0].plot(prob_dist)
axes[0].set_xlabel('No. of customers arriving in 4 minutes')
axes[0].set_ylabel('Probability')
axes[0].set_title('Poisson Distribution: mu=4.8')
axes[1].plot(cum_dist)
axes[1].set_xlabel('No. of customers arriving in 4 minutes')
axes[1].set_ylabel('Probability')
axes[1].set_title('Cumulative Poisson Distribution: mu=4.8')
plt.show()
```

## Problem Statement 10:
Please compute the following:

a) P(Z > 1.26), . P(Z < ‚àí0.86), P(Z > ‚àí1.37), P(‚àí1.25 < Z < 0.37), . P(Z ‚â§ ‚àí4.6)

b) Find the value ùëß such that ùëÉ(ùëç > ùëß) = 0.05

c) Find the value of ùëß such that ùëÉ(‚àíùëß < ùëç < ùëß) = 0.99

```python
from scipy.stats import norm

# P(Z > 1.26)
prob1 = 1 - norm.cdf(1.26)
print('P(Z > 1.26) = ', round(prob1, 4))

# P(Z < ‚àí0.86)
prob2 = norm.cdf(-0.86)
print('P(Z < ‚àí0.86) = ', round(prob2, 4))

# P(Z > ‚àí1.37)
prob3 = 1 - norm.cdf(-1.37)
print('P(Z > ‚àí1.37) = ', round(prob3, 4))

# P(‚àí1.25 < Z < 0.37)
prob4 = norm.cdf(0.37) - norm.cdf(-1.25)
print('P(‚àí1.25 < Z < 0.37) = ', round(prob4, 4))

# P(Z ‚â§ ‚àí4.6)
prob5 = norm.cdf(-4.6)
print('P(Z ‚â§ -4.6) = ', round(prob5, 4))

# The value ùëß such that ùëÉ(ùëç > ùëß) = 0.05
z = norm.isf(0.05)
print('The value ùëß such that ùëÉ(ùëç > ùëß) = 0.05 : ', round(z, 4))

# The value of ùëß such that ùëÉ(‚àíùëß < ùëç < ùëß) = 0.99
z = norm.isf((1 - 0.99)/2)
print('The value of ùëß such that ùëÉ(‚àíùëß < ùëç < ùëß) = 0.99 : ', round(z, 4))
```

## Problem Statement 11:
The current flow in a copper wire follow a normal distribution with a mean of 10 ùëöA and a variance of 4 (ùëöùê¥)^2.
What is the probability that a current measurement will exceed 13 ùëöùê¥? What is the probability that a current measurement is
between 9 and 11mA? Determine the current measurement which has a probability of 0.98.

```python
# mu = 10, var = 4 , std = 2

# Probability that a current measurement will exceed 13 ùëöùê¥
prob1 = norm.sf(x = 13, loc = 10, scale = 2)
print('Probability that a current measurement will exceed 13 ùëöùê¥ : ', round(prob1, 4))

# Probability that a current measurement is between 9 and 11 mA
prob2 = norm.cdf(x = 11, loc = 10, scale = 2) - norm.cdf(x = 9, loc = 10, scale = 2)
print('Probability that a current measurement is between 9 and 11 mA : ', round(prob2, 4))

# Current measurement which has a probability of 0.98
i = norm.ppf(q = 0.98, loc = 10, scale = 2)
print('Current measurement i, which has a probability of P(I < i) = 0.98 : i =', round(i, 4), 'mA')
```

The specifications of the shaft are 0.2500 ‚àì 0.0015 inch. What proportion of shafts are in sync with the specifications?
If the process is centered so that the mean is equal to the target value of 0.2500, what proportion of shafts conform to the
new specifications? What is your conclusion from this experiment?

```python
# mean = 0.2508 and sd = 0.0005
# Z = (X - mean)/sd = (X - 0.2508)/0.0005
# P(0.2485 < X < 0.2515) = P((0.2485-0.2508)/0.0005 < Z < (0.2515-0.2508)/0.0005) = P(‚àí4.6 < Z < 1.4)
prob1 = norm.cdf(1.4) - norm.cdf(-4.6)
print('P(0.2485 < X < 0.2515) = P(‚àí4.6 < Z < 1.4) = ', round(prob1, 4))

# mean = 0.2500 and sd = 0.0005
# Z = (X - mean)/sd = (X - 0.2500)/0.0005
# P(0.2485 < X < 0.2515) = P((0.2485-0.2500)/0.0005 < Z < (0.2515-0.2500)/0.0005) = P(‚àí3.0 < Z < 3.0)
# Thus the diameters of the shaft fall within 3 standard deviations on either side from the mean 
prob2 = norm.cdf(3.0) - norm.cdf(-3.0)
print('P(0.2485 < X < 0.2515) = P(‚àí3.0 < Z < 3.0) = ', round(prob2, 4))
```
