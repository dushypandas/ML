```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
iris.target_names
```

```python
# iris data
x = iris.data
y = iris.target

# transform iris data into 3 dimensions
pca = decomposition.PCA(n_components=3)
X = pca.fit_transform(x)
```

```python
x.shape , X.shape
```

```python
# 3d chart with transformed dimensions

fig = plt.figure(figsize=(8, 8))

ax = Axes3D(fig)
ax.set_xlabel('First Principal Component', fontsize = 15)
ax.set_ylabel('Second Principal Component', fontsize = 15)
ax.set_zlabel('Third Principal Component', fontsize = 15)
ax.set_title('Three Dimensions of Iris plants dataset', fontsize = 18)

for color, value in [('r', 0), ('g', 1), ('b', 2)]:
    ax.scatter(X[y == value, 0], X[y == value, 1], X[y == value, 2], c=color, s=50)

ax.legend(['Setosa', 'Versicolour', 'Virginica'])
```
